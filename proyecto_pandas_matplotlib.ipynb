# **Análisis de datos: Libros Superventas**
**Usando Pandas, Numpy, Matplotlib y Seaborn.**



Paso 1: Llamar las librerias que se van a trabajar

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Paso 2: Leer el archivo

```python
bestsellers = pd.read_csv('/content/bestsellers-with-categories.csv')
bestsellers
```

###Limpieza

Paso 3: Identificar registros nulos

```python
bestsellers.isnull().sum()
```

Paso 4: Borrar datos nulos

```python
bestsellers.dropna()
```

## Construcción del Dashboard:

### Reviews por año

```python
reviews= sns.lineplot(data=bestsellers, x='Year', y='Reviews')
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=bestsellers, x='Year', y='Reviews', palette=sns.blend_palette(['#55E3EC','#6FABD3']))
#plt.xticks(rotation=45)
plt.xlabel('Year')
plt.ylabel('Reviews')
plt.title('Reviews by Year')
plt.show()
```

###Libros que se publicaron en una fecha inferior al 2017

```python
me_2017 = bestsellers['Year'] < 2017
bestsellers[me_2017]
```

###10 bestsellers publicados en el 2018



```python
y2018 = bestsellers['Year'] == 2018
best10 = bestsellers[y2018].head(10)
best10
```

###Género que obtuvo más reseñas (No ficción y ficción)

```python
plt.figure(figsize=(8,5))
sns.countplot(data=bestsellers, x='Genre', palette=['#55E3EC','#6FABD3'])
plt.xticks(np.arange(2),['Non Fiction','Fiction'])
plt.xlabel('Genre')
plt.ylabel('Amount of books')
plt.title('Gender that got the most reviews')
plt.show()
```
